### Import the library and dataset

```python
import pandas as pd
import plotly.express as px
from dash import dash, html, dcc
from dash.dependencies import Input, Output
```

```python
def importdata() :
    var = pd.read_csv('airline_data.csv')
    return var

if __name__ == '__main__':
    airline_data = importdata()
```

### Create a dropdown menu

The code below will create a dropdown menu for a list of cities

```python
# Create dash application
app = dash.Dash(__name__)

app.layout = html.Div(children=[
    html.H1('DashBoard',style={'text-align':'center','color':'#503D36','fontSize':40}),
    html.Div([
        dcc.Dropdown(options=[
            {'label':'New York City', 'value':'NYC'},
            {'label':'Montreal','value':'MTL'},
            {'label':'San Francisco','value':'SF'}],
                    value='NYC')]),
    ])

# Run application
if __name__ == '__main__' :
    app.run_server()
```

The code below will create a bargraph dashboard with the city as the input

```python
# Add Dataframe
df = pd.DataFrame({
    "Fruit": ["Apples", "Oranges", "Bananas", "Apples", "Oranges", "Bananas"],
    "Amount": [4, 1, 2, 2, 4, 5],
    "City": ["SF", "SF", "SF", "NYC", "MTL", "NYC"]})
df
```

```python
# Create dash application
app = dash.Dash(__name__)

# Create html layout 
app.layout = html.Div(children=[
    html.H1('Dashboard',style={'text-align':'center','color':'#503D36','fontSize':40}),
    html.Div(dcc.Dropdown(id='city-code',
                          options=[
                              {'label':'New York City','value':'NYC'},
                              {'label':'Montreal','value':'MTL'},
                              {'label':'San Francisco','value':'SF'}
                          ])),
    html.Br(),
    html.Br(),
    html.Div(dcc.Graph(id='bar-fig'))
])

@app.callback(
    Output('bar-fig','figure'),
    Input('city-code','value')
)

def get_graph(city_code) :
    # Obtain the dataframe for the specified city
    df_bar = df[df['City']==str(city_code)]
    # Plot the dataframe
    fig = px.bar(df_bar,x='Fruit',y='Amount')
    
    return fig

# Run the application
if __name__ == '__main__':
    app.run_server()
```

# Car Automobile Practice Exercise Dashboard

```python
def automobileimport():
    var = pd.read_csv('automobileEDA.csv')
    return var
if __name__ == '__main__':
    auto_data = automobileimport()

auto_data.head()
```

Using the dataset above, create a dashboard consisting of a pie chart and a bar chart that display the relationship between
body-style and price, filtered using the drive wheel type (dropdown).

```python
auto_data['drive-wheels'].unique()
```

```python
# Create the app
app = dash.Dash(__name__)

# Create the html layout
app.layout = html.Div(children=[
    html.H1('Car Automobile Components',style={'text-align':'center','fontSize':40,'color':'#503D36'}),
    html.Br(),
    html.H2('Drive Wheel Type'),
    html.Div(dcc.Dropdown(id='wheel-type',options=[
        {'label':'Rear Wheel Drive','value':'rwd'},
        {'label':'Front Wheel Drive','value':'fwd'},
        {'label':'4 Wheel Drive','value':'4wd'}]
                         )),
    html.Br(),
    html.Div([
        html.Div(dcc.Graph(id='pie_fig')),
        html.Div(dcc.Graph(id='bar_fig'))
    ],style={'display':'flex'})
])

# Create app callback
@app.callback([
    Output('pie_fig','figure'),
    Output('bar_fig','figure')
],Input('wheel-type','value'))

def get_graph(wheel_type):
    # Get the data for the chosen wheel type
    df = auto_data[auto_data['drive-wheels']==str(wheel_type)].groupby('body-style',as_index=False)[['price']].mean()
    # Plot the pie chart
    pie_fig = px.pie(df,values='price',names='body-style',title='Pie Chart')
    # Plot the bar chart
    bar_fig = px.bar(df,x='body-style',y='price',title='Bar Chart')
    
    return [pie_fig, bar_fig]

# Run the application
if __name__ == '__main__':
    app.run_server()
```
